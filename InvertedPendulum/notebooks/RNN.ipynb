```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import clear_output

import torch
import torch.nn as nn
import torch.optim as optim
```

```python
states = np.load('../data/states.npy')[:6000]
forces = np.load('../data/forces.npy')[:5999]
```

states = [X, angle, Xdot, angledot]

```python
print("states shape:",states.shape)
print("forces shape:", forces.shape)
```

```python
sin_angle = np.sin(states[:,1])
cos_angle = np.cos(states[:,1])

X = np.concatenate((states[:-1,0][:,np.newaxis],sin_angle[:-1,np.newaxis],cos_angle[:-1,np.newaxis],states[:-1,2:],forces[:,np.newaxis]),axis=1)

mean = np.mean(X,axis=0)
std = np.std(X,axis=0)

y = np.concatenate((states[1:,0][:,np.newaxis],sin_angle[1:,np.newaxis],cos_angle[1:,np.newaxis],states[1:,2:]),axis=1)

X = (X-mean)/std
y = (y-mean[:-1])/std[:-1]

print("X shape:",X.shape)
print("y shape:", y.shape)
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(range(X.shape[0]),X,alpha=0.6)
plt.show()
```

```python
X = torch.Tensor(X)
y = torch.Tensor(y)
```

```python
TEST_RATIO = 0.2
BATCH = 10
m,n = X.shape

batch_size = m//BATCH

X_train_dataloader = []
y_train_dataloader = []
X_test_dataloader = []
y_test_dataloader = []

for batch in range(BATCH):
    X_train_dataloader.append(X[batch*batch_size:(batch+1)*batch_size-int(TEST_RATIO*batch_size)])
    y_train_dataloader.append(y[batch*batch_size:(batch+1)*batch_size-int(TEST_RATIO*batch_size)])
    X_test_dataloader.append(X[(batch+1)*batch_size-int(TEST_RATIO*batch_size):(batch+1)*batch_size])
    y_test_dataloader.append(y[(batch+1)*batch_size-int(TEST_RATIO*batch_size):(batch+1)*batch_size])
    
```

```python
class Net(nn.Module):
    def __init__(self):
        self.INPUT_FEATURES = 6
        self.OUTPUT_FEATURES = 5
        super().__init__()
        self.fc_1 = nn.Sequential(nn.Linear(self.INPUT_FEATURES, 10),nn.LeakyReLU(),
                                 nn.Linear(10,10),nn.LeakyReLU(),
                                 nn.Linear(10,self.OUTPUT_FEATURES))

    def forward(self, x):
        x = self.fc_1(x)
        return x
```

```python
net = Net().cuda()
loss_function = nn.MSELoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)
```

```python
net.train()

EPOCH = 10000

for epoch in range(EPOCH):
    for batch in range(BATCH):
        X_batch = X_train_dataloader[batch].cuda()
        y_batch = y_train_dataloader[batch].cuda()
        
        output = net(X_batch)
        loss = loss_function(output,y_batch)
        loss.backward()
        optimizer.step()
    clear_output(wait=True)
    print(epoch)
```

```python
fig, ax = plt.subplots(5,1,figsize=(30,10))

net.eval()
with torch.no_grad():
    X_cuda = X.cuda()
    output = net(X_cuda)
    output = output.cpu()
    clear_output(wait=True)
for feature in range(y.shape[1]):
    ax[feature].plot(range(output.shape[0]),y[:,feature],label='actual')
    ax[feature].plot(range(output.shape[0]),output[:,feature],label='generated')
    ax[feature].plot(range(output.shape[0]),X[:,-1],label='force',alpha=0.3)
    ax[feature].legend()
    ax[feature].set_ylim([-10,10])
```

```python
X_initial =  np.array([-0.03,0.0,1.0,0.0,0.0])
```

```python
output = X_initial
outputs = [output]

STEPS = 100

net.eval()
with torch.no_grad():
    for _ in range(STEPS):
        output = np.append(output,0)
        output = torch.Tensor(output).cuda()
        output = net(output)
        output = output.cpu()
        outputs.append(np.array(output))
```
